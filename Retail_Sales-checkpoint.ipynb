# New Column

**Cost Price → Profit**
- Price per Unit (Selling Price)
- Total Amount (Revenue)
- Unit Per Cost(add)
- total Cost(add)
- Profit(add)

```python
df['Cost Per Unit'] = df['Price per Unit'] * np.random.uniform(0.6, 0.9, len(df))
```

```python
df["Total Cost"]= df["Cost Per Unit"]*df["Quantity"]
```

```python
df["Profit Per Unit"] = df["Price per Unit"]-df["Cost Per Unit"]
```

```python
df["Total Profit"] = df["Profit Per Unit"]*df["Quantity"]
```

**ADD loctaion columns**

```python
df['Location'] = np.random.choice(
    ['Mumbai', 'Delhi', 'Bangalore', 'Chennai'],
    size=len(df),
    p=[0.4, 0.3, 0.2, 0.1]
)
```

**Customer Type**
- Example:High spender/Low spender

```python
df["Customer Type"] = df["Total Amount"].apply(lambda x:'High Spender' if x>=1000 else 'Low Spender')
```

# data preprocessing 

**Main Business Problem:**
- “How can a retail company increase its sales and profit, and better understand its customers?”

**Customer Understanding Problem**

```python
total_sales= df.groupby("Gender")["Total Amount"].sum().reset_index()
total_sales
```

```python
sns.barplot(
    x="Gender",
    y="Total Amount",
    data=total_sales,   
)
plt.title("Total ravenue by Gender", fontsize=10)
plt.xlabel("Gender")
plt.ylabel("Total revenue")

plt.show()
```

```python
total_profit= df.groupby("Gender")["Total Profit"].sum().reset_index()
total_profit
```

```python
sns.barplot(
    x="Gender",
    y="Total Profit",
    data=total_profit,   
)
plt.title("Total profit by Gender", fontsize=10)
plt.xlabel("Gender")
plt.ylabel("Total profit")

plt.show()
```

```python
df_compare = df.groupby("Gender").agg({
    "Total Amount": "sum",
    "Total Profit": "sum"
}).reset_index()
df_compare
```

```python
df_melt = df_compare.melt(id_vars="Gender",
                          value_vars=["Total Amount", "Total Profit"],
                          var_name="Metric", value_name="Value")

plt.figure()

sns.barplot(x="Gender", y="Value", hue="Metric", data=df_melt)

plt.title("Gender-wise Revenue vs Profit")
plt.show()
```

**Product Performance Problem**

```python
# Group data
category_revenue = df.groupby(["Product Category"])["Total Amount"].sum().reset_index()
category_revenue
```

```python
# Plot
sns.barplot(x="Total Amount", y="Product Category", data=category_revenue)

plt.title("Revenue by Product Category")
plt.xlabel("Total Revenue")
plt.ylabel("Category")

plt.show()
```

```python
category_profit = df.groupby("Product Category")["Total Profit"].sum().reset_index()

sns.barplot(x="Total Profit", y="Product Category", data=category_profit)

plt.title("Profit by Product Category")
plt.show()
```

**Pro Insight:**
- High Revenue + Low Profit → reduce cost or increase price
- Low Revenue + High Profit → increase marketing 

**Revenue Optimization Problem**

```python
df_compare = df.groupby("Product Category").agg({
    "Total Amount": "sum",
    "Total Profit": "sum"
}).reset_index()
df_compare
```

```python
df_melt = df_compare.melt(id_vars="Product Category", 
                          value_vars=["Total Amount", "Total Profit"],
                          var_name="Metric", value_name="Value")

sns.barplot(x="Product Category", y="Value", hue="Metric", data=df_melt)

plt.title("Revenue vs Profit by Category")
plt.show()
```

**Sales Trend Problem**

```python
df["Month"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["Total Amount"].sum().reset_index()
monthly_sales["Month"] = monthly_sales["Month"].astype(str)
```

```python
plt.figure()

sns.lineplot(x="Month", y="Total Amount", data=monthly_sales, marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)

plt.show()
```

```python
monthly_data = df.groupby("Month").agg({
    "Total Amount": "sum",
    "Total Profit": "sum"
}).reset_index()

monthly_data["Month"] = monthly_data["Month"].astype(str)

# melt for comparison
df_melt = monthly_data.melt(id_vars="Month", 
                           value_vars=["Total Amount", "Total Profit"],
                           var_name="Metric", value_name="Value")

plt.figure()
sns.lineplot(x="Month", y="Value", hue="Metric", data=df_melt, marker="o")

plt.title("Sales vs Profit Trend")
plt.xticks(rotation=90)
plt.show()
```

**Customer Segmentation Problem**

```python
total_sales_customer_type = df.groupby(
    ["Gender", "Customer Type"]
)["Total Amount"].sum().unstack()
total_sales_customer_type
```

```python
total_sales_customer_type.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Total Sales Amount")
plt.title("Sales by Gender and Customer Type")
plt.xticks(rotation=0)

plt.show()
```

```python
total_profit_customer_type = df.groupby(
    ["Gender", "Customer Type"]
)["Total Profit"].sum().unstack()
total_profit_customer_type
```

```python
total_profit_customer_type.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Total Sales Amount")
plt.title("Sales by Gender and Customer Type")
plt.xticks(rotation=0)

plt.show()
```

**Pro Insight:**
- sales + female  → male 
- Transaction Count  + low spender >high spender(transaction)  
- total sale + High spender → low spender(sales)
